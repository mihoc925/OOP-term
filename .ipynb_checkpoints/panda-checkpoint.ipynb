```python
import pandas as pd
df = pd.read_csv('api/Weather.csv', sep=',')
# df['datetime'] = df['date'].apply(lambda x: pd.to_datetime(str(x), format='%Y-%m-%d'))
# df['month'] = pd.DatetimeIndex(df['date']).month
df['year'] = pd.DatetimeIndex(df['date']).year
df.loc[[0,1,9]]
```

```python
import matplotlib.pyplot as plt
data1 = df.loc[df['province']== 'Seoul']
data1grp = data1.groupby('year')[['avg_temp', 'min_temp', 'max_temp']].mean()
# print(data1grp)
data1plot = data1grp.plot()
data1plot.set_title("Seoul")

data2 = df.loc[df['province']== 'Busan']
data2grp = data2.groupby('year')[['avg_temp', 'min_temp', 'max_temp']].mean()
# print(data2grp)
data2plot = data2grp.plot()
data2plot.set_title("Busan")

data3 = df.loc[df['province']== 'Chungcheongbuk-do']
data3grp = data3.groupby('year')[['avg_temp', 'min_temp', 'max_temp']].mean()
# print(data3grp)
data3plot = data3grp.plot()
data3plot.set_title("Chungcheongbuk-do")
```
